```python

# Lista de inscritos
nombres = [
    "Alice", "Bob", "Charlie", "David", "Eve",
    "Frank", "Grace", "Heidi", "Ivan", "Judy"
]

# lista Categorías
categorias = ["5k", "10k", "21k"]

# Crear la lista de diccionarios
inscritos = [
    {"nombre": nombre, "categoria": random.choice(categorias)}
    for nombre in nombres
]

# Función para mostrar los inscritos
def mostrar_inscritos(inscritos):
    for inscrito in inscritos:
        print(f"Nombre: {inscrito['nombre']}, Categoría: {inscrito['categoria']}")

# mostrar datos
mostrar_inscritos(inscritos)
```

```python



productos = {
    "pan": {"precio": 5000, "cantidad": 1},
    "leche": {"precio": 4000, "cantidad": 5},
    "huevos": {"precio": 20000, "cantidad": 2}
}

# Calculamos el subtotal para cada producto y lo añadimos al diccionario
for nombre, info in productos.items():
    info["subtotal"] = info["precio"] * info["cantidad"]

# Imprimimos la tabla
print(f"{'Producto':<10} {'Precio':<10} {'Cantidad':<10} {'Subtotal':<10}")
print("-" * 40)
for nombre, info in productos.items():
    print(f"{nombre:<10} {info['precio']:<10} {info['cantidad']:<10} {info['subtotal']:<10}")

# Calculamos el total de la factura
total = sum(info["subtotal"] for info in productos.values())
print("-" * 40)
print(f"{'Total':<30} {total:<10}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Tus listas de datos
value_n_list = ['pan', 'leche', 'huevos', 'cheetos', 'papas', 'yogurt', 'gaseosa', 'bombillo', 'galletas', 'chocoramo']
value_c_list = [2, 3, 4, 1, 10, 1, 2, 3, 4, 5]
value_p_list = [1000, 3000, 4000, 16000, 1000, 1000, 1000, 1000, 1000, 1000]

# Diccionario de datos
data_dict = {
    "Producto": value_n_list,
    "Cantidad": value_c_list,
    "Precio": value_p_list,
}

# Crear DataFrame
df = pd.DataFrame(data_dict)

# Calcular la matriz de correlación
#correlation_matrix = df[['Cantidad', 'Precio']].corr()
correlation_matrix = df.corr()
# Mostrar la matriz de correlación
print("Matriz de Correlación:")
print(correlation_matrix)

# Visualizar la matriz de correlación con un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```
