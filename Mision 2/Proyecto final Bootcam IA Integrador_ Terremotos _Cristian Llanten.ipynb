Limpia columnas irrelevantes para la clasificación

```python
columns_to_remove = ['id','url', 'detailUrl','place', 'location', 'geometryType', 'what3words']
df = df.drop(columns=columns_to_remove)
df.head()
```

Transformación de columnas categóricas
Codificado las columnas de texto para que sean numéricas usando LabelEncoder.

```python
# Codificar columnas categóricas
label_enc = LabelEncoder()
categorical_columns = ['type', 'alert', 'status', 'net', 'magType', 'continent', 'country', 'subnational', 'city']

for col in categorical_columns:
    df[col] = label_enc.fit_transform(df[col].astype(str))
```

Selección de variables y separación de datos
Define tus variables X (características) y y (objetivo) y divide los datos en conjuntos de entrenamiento y prueba.

```python
# Seleccionar variables (X) y objetivo (y)
features_to_keep = ['magnitude', 'distanceKM', 'latitude', 'longitude']
X = df[features_to_keep]
y = df['type']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```
