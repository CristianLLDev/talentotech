```python
def separar_pares_impares(inicio, fin):
    pares = []
    impares = []
    for num in range(inicio, fin+1):
        if num % 2 == 0:
            pares.append(num)
        else:
            impares.append(num)
    return pares, impares


def imprimir_lista(nombre_lista, lista):
    print(f"Lista de {nombre_lista}: {lista}")


pares, impares = separar_pares_impares(1, 40)


imprimir_lista("números pares", pares)
imprimir_lista("números impares", impares)
```
