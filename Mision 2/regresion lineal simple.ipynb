```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear las listas de datos
x = [1, 2, 2, 3, 4, 4, 5, 6]
y = [2, 3, 4, 4, 4, 6, 5, 7]

# Crear un diccionario con las listas
data = {
    'X': x,
    'Y': y
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data)

# Crear una nueva columna con el producto de X e Y
df['X * Y'] = df['X'] * df['Y']

# Crear otra columna con el valor de X al cuadrado
df['X^2'] = df['X'] ** 2

# Calcular las sumatorias
sum_x = df['X'].sum()
sum_y = df['Y'].sum()
sum_x_y = df['X * Y'].sum()
sum_x_squared = df['X^2'].sum()

# Añadir las sumatorias como una nueva fila al DataFrame
df.loc['Total'] = [sum_x, sum_y, sum_x_y, sum_x_squared]

# Mostrar el DataFrame con las sumatorias
print("DataFrame:")
print(df)

# Calcular los coeficientes de la regresión lineal
n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)

# Cálculo de la pendiente (m)
numerator = sum_x_y - n * x_mean * y_mean
denominator = sum_x_squared - n * x_mean**2
m = numerator / denominator

# Cálculo de la intersección (b)
b = y_mean - m * x_mean

# Mostrar los coeficientes
print("\nCoeficientes de la regresión lineal:")
print(f"Pendiente (m): {m:.2f}")
print(f"Intersección (b): {b:.2f}")

# Graficar los datos y la línea de regresión
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Datos')#dispersion
plt.plot(x, [m * xi + b for xi in x], color='blue', label='Línea de Regresión')

# Añadir etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de Y vs X con Línea de Regresión')
plt.legend()
plt.grid(True)

# Mostrar la gráfica
plt.show()

# Calcular los residuos y S_r (sumatoria de residuos)
y_pred = [m * xi + b for xi in x]  # Predicciones de la regresión
residuos = np.array(y) - np.array(y_pred)  # Residuos
Sr = np.sum(residuos**2)  # Sumatoria de los residuos cuadrados

# Mostrar el valor de S_r
print(f"\nSumatoria de los residuos (S_r): {Sr:.2f}")
```
