```python
# Genera estadísticas descriptivas del DataFrame, como la media, la desviación estándar, los valores mínimo y máximo, etc.
df.describe()
```

```python
# Crea un histograma para una columna específica del DataFrame.
df['magnitude'].hist()
plt.xlabel('Magnitud')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Magnitud de los Terremotos')
plt.show()
```

```python
# Crea un diagrama de dispersión para dos columnas específicas del DataFrame.
df.plot(kind='scatter', x='longitude', y='latitude', alpha=0.4)
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Ubicación de los Terremotos')
plt.show()
```

```python
# Crea una matriz de correlación para todas las columnas numéricas del DataFrame.
df['id'] = df['id'].astype(float)
df.corr()
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
