```python
import pandas as pd
import numpy as np

# Crear un DataFrame con datos simulados
data = {
    'equipo_local': [
        'EquipoA', 'EquipoB', 'EquipoC', 'EquipoD', 'EquipoE',
        'EquipoA', 'EquipoB', 'EquipoC', 'EquipoD', 'EquipoE',
        'EquipoA', 'EquipoB', 'EquipoC', 'EquipoD', 'EquipoE',
        'EquipoA', 'EquipoB', 'EquipoC', 'EquipoD', 'EquipoE'
    ],
    'equipo_visitante': [
        'EquipoB', 'EquipoC', 'EquipoD', 'EquipoE', 'EquipoA',
        'EquipoC', 'EquipoD', 'EquipoE', 'EquipoA', 'EquipoB',
        'EquipoD', 'EquipoE', 'EquipoA', 'EquipoB', 'EquipoC',
        'EquipoE', 'EquipoA', 'EquipoB', 'EquipoC', 'EquipoD'
    ],
    'resultado': [
        0, 1, 2, 1, 0,  # 0 = derrota del local, 1 = empate, 2 = victoria del local
        1, 2, 0, 2, 1,
        0, 1, 2, 1, 0,
        2, 0, 1, 0, 2
    ]
}

df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_partidos.csv', index=False)
print("Base de datos creada y guardada en 'datos_partidos.csv'")
data
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Cargar los datos
data = pd.read_csv('datos_partidos.csv')

# Codificación de variables categóricas
label_encoder = LabelEncoder()
data['equipo_local'] = label_encoder.fit_transform(data['equipo_local'])
data['equipo_visitante'] = label_encoder.transform(data['equipo_visitante'])

# Definir características y etiqueta
X = data[['equipo_local', 'equipo_visitante']]
y = data['resultado']

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalizar características
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Crear el modelo
model = LogisticRegression(max_iter=1000)

# Entrenar el modelo
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Ejemplo de datos de un partido futuro
nuevo_partido = pd.DataFrame({
    'equipo_local': [label_encoder.transform(['EquipoB'])],
    'equipo_visitante': [label_encoder.transform(['EquipoC'])]
})
print(nuevo_partido)
# Normalizar los datos del nuevo partido
nuevo_partido = scaler.transform(nuevo_partido)

# Predecir el resultado
resultado_predicho = model.predict(nuevo_partido)
print(f'Resultado Predicho: {resultado_predicho[0]}')
```
