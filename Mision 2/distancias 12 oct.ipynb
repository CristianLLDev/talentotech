```python
import math

x1=1
x2=4
y1=2
y2=6
def distancia_euclidiana(x1, y1, x2, y2):
 distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
 return distancia

resultado = distancia_euclidiana(x1, y1, x2, y2)

# Print the result
print(resultado)
```

```python
# Calcular la distancia entre los puntos (1, 2) y (4, 6)
distancia = distancia_euclidiana(1, 2, 4, 6)
print(distancia)  # Salida: 5.0
```

```python
import matplotlib.pyplot as plt
import math

x1 = 1
x2 = 4
y1 = 2
y2 = 6

# Calculate Euclidean distance
distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Create the plot
plt.plot([x1, x2], [y1, y2], marker='o', color='blue')  # Plot points and line
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Euclidean Distance")
plt.grid(True)  # Show grid

# Annotate points
plt.annotate(f"({x1}, {y1})", (x1, y1), textcoords="offset points", xytext=(0, 10), ha='center')
plt.annotate(f"({x2}, {y2})", (x2, y2), textcoords="offset points", xytext=(0, 10), ha='center')

# Show the plot
plt.show()
```

```python
def distancia_manhattan(x1, y1, x2, y2):
    distancia = abs(x2 - x1) + abs(y2 - y1)
    return distancia
```

```python
resultado = distancia_manhattan(x1, y1, x2, y2)

# Print the result
print(resultado)
```
