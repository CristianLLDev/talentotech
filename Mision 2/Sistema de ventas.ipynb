```python
import pandas as pd
import sqlite3

# Conectar a la base de datos SQLite (se crearÃ¡ si no existe)
conn = sqlite3.connect('ventas.db')
c = conn.cursor()

# Crear una tabla si no existe
c.execute('''
CREATE TABLE IF NOT EXISTS ventas (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    producto TEXT,
    cantidad INTEGER,
    precio REAL,
    subtotal REAL
)
''')

# FunciÃ³n para agregar una venta
def agregar_venta(producto, cantidad, precio):
    subtotal = cantidad * precio
    c.execute('''
    INSERT INTO ventas (producto, cantidad, precio, subtotal)
    VALUES (?, ?, ?, ?)
    ''', (producto, cantidad, precio, subtotal))
    conn.commit()

# Ejemplo de ingreso de datos
print("Ingrese los datos de la venta:")
producto = input("Producto: ")
cantidad = int(input("Cantidad: "))
precio = float(input("Precio: "))
agregar_venta(producto, cantidad, precio)

print("Venta agregada con Ã©xito.")#

# Leer los datos desde la base de datos
df = pd.read_sql_query('SELECT * FROM ventas', conn)

# Mostrar el DataFrame
print(df)

# Visualizar los datos
plt.figure(figsize=(10, 6))

# GrÃ¡fico de ventas por producto
df.groupby('producto')['subtotal'].sum().plot(kind='bar')
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Subtotal')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()

# Cerrar la conexiÃ³n
conn.close()
```
